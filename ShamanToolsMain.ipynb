```python
# Estructuras de datos iniciales
inventario = [
    {"ID": 1, "Material": "Ladrillo 10x20x30", "PrecioCompra": 1200, "PrecioVenta": 1500, "CantidadDisponible": 1000, "TipoUnidad": "un"},
    {"ID": 2, "Material": "Cemento Gris x50kg", "PrecioCompra": 28000, "PrecioVenta": 30000, "CantidadDisponible": 200, "TipoUnidad": "saco"},
    {"ID": 3, "Material": "Varilla Corrugada #5 x6m", "PrecioCompra": 5000, "PrecioVenta": 5500, "CantidadDisponible": 150, "TipoUnidad": "un"},
    {"ID": 4, "Material": "Impermeabilizante 22kg", "PrecioCompra": 350000, "PrecioVenta": 360000, "CantidadDisponible": 50, "TipoUnidad": "un"},
    {"ID": 5, "Material": "Malla Electro Soldada 6x2.2m", "PrecioCompra": 42000, "PrecioVenta": 45000, "CantidadDisponible": 75, "TipoUnidad": "un"},
]

usuarios = [
    {"ID": 1, "Usuario": "ivan", "Contraseña": "132512", "Rol": "admin"},
    {"ID": 2, "Usuario": "brina", "Contraseña": "111222333", "Rol": "vendedor"},
    {"ID": 3, "Usuario": "maria camila", "Contraseña": "201212", "Rol": "vendedor"},
]

# Funciones del sistema
def registrar_usuario(nombre, contraseña, rol):
    nuevo_usuario = {"ID": len(usuarios) + 1, "Usuario": nombre, "Contraseña": contraseña, "Rol": rol}
    usuarios.append(nuevo_usuario)
    print(f"Usuario {nombre} registrado con éxito.")

def agregar_producto(id, material, precio_compra, precio_venta, cantidad, tipo_unidad):
    nuevo_producto = {
        "ID": id,
        "Material": material,
        "PrecioCompra": precio_compra,
        "PrecioVenta": precio_venta,
        "CantidadDisponible": cantidad,
        "TipoUnidad": tipo_unidad
    }
    inventario.append(nuevo_producto)
    print(f"Producto {material} agregado al inventario.")

def actualizar_producto(id, precio_compra, precio_venta, cantidad, tipo_unidad):
    for producto in inventario:
        if producto["ID"] == id:
            producto["PrecioCompra"] = precio_compra
            producto["PrecioVenta"] = precio_venta
            producto["CantidadDisponible"] = cantidad
            producto["TipoUnidad"] = tipo_unidad
            print(f"Producto {producto['Material']} actualizado.")
            return
    print("Producto no encontrado.")

def consultar_inventario():
    print("Inventario de productos:")
    for producto in inventario:
        print(f"ID: {producto['ID']}, Material: {producto['Material']}, Cantidad Disponible: {producto['CantidadDisponible']}")

def comprar_producto(id, cantidad):
    for producto in inventario:
        if producto["ID"] == id:
            producto["CantidadDisponible"] += cantidad
            print(f"Se han comprado {cantidad} unidades de {producto['Material']}.")
            return
    print("Producto no encontrado. Agregue el producto primero.")

def vender_producto(id, cantidad):
    for producto in inventario:
        if producto["ID"] == id:
            if producto["CantidadDisponible"] >= cantidad:
                producto["CantidadDisponible"] -= cantidad
                print(f"Se han vendido {cantidad} unidades de {producto['Material']}.")
            else:
                print("Cantidad insuficiente en inventario.")
            return
    print("Producto no encontrado.")

# Función principal para iniciar el sistema
def iniciar_sistema():
    usuario = input("Ingrese usuario: ")
    contraseña = input("Ingrese contraseña: ")

    usuario_autenticado = next((u for u in usuarios if u["Usuario"] == usuario and u["Contraseña"] == contraseña), None)

    if usuario_autenticado:
        print(f"Bienvenido, {usuario_autenticado['Usuario']} ({usuario_autenticado['Rol']})")
        while True:
            if usuario_autenticado["Rol"] == "admin":
                print("\nOpciones (admin):")
                print("1. Registrar usuario")
                print("2. Agregar producto")
                print("3. Actualizar producto")
                print("4. Consultar inventario")
                print("5. Comprar producto")
                print("6. Salir")
                opcion = input("Seleccione una opción: ")

                if opcion == "1":
                    nombre = input("Nombre de usuario: ")
                    contraseña = input("Contraseña: ")
                    rol = input("Rol (admin/vendedor): ")
                    registrar_usuario(nombre, contraseña, rol)
                elif opcion == "2":
                    id = int(input("ID del producto: "))
                    material = input("Nombre del material: ")
                    precio_compra = float(input("Precio de compra: "))
                    precio_venta = float(input("Precio de venta: "))
                    cantidad = int(input("Cantidad disponible: "))
                    tipo_unidad = input("Tipo de unidad: ")
                    agregar_producto(id, material, precio_compra, precio_venta, cantidad, tipo_unidad)
                elif opcion == "3":
                    id = int(input("ID del producto a actualizar: "))
                    precio_compra = float(input("Nuevo precio de compra: "))
                    precio_venta = float(input("Nuevo precio de venta: "))
                    cantidad = int(input("Nueva cantidad disponible: "))
                    tipo_unidad = input("Nuevo tipo de unidad: ")
                    actualizar_producto(id, precio_compra, precio_venta, cantidad, tipo_unidad)
                elif opcion == "4":
                    consultar_inventario()
                elif opcion == "5":
                    id = int(input("ID del producto a comprar: "))
                    cantidad = int(input("Cantidad a comprar: "))
                    comprar_producto(id, cantidad)
                elif opcion == "6":
                    print("Saliendo del sistema...")
                    break
                else:
                    print("Opción no válida.")

            elif usuario_autenticado["Rol"] == "vendedor":
                print("\nOpciones (vendedor):")
                print("1. Consultar inventario")
                print("2. Vender producto")
                print("3. Salir")
                opcion = input("Seleccione una opción: ")

                if opcion == "1":
                    consultar_inventario()
                elif opcion == "2":
                    id = int(input("ID del producto a vender: "))
                    cantidad = int(input("Cantidad a vender: "))
                    vender_producto(id, cantidad)
                elif opcion == "3":
                    print("Saliendo del sistema...")
                    break
                else:
                    print("Opción no válida.")
    else:
        print("Usuario o contraseña incorrectos.")

# Ejecutar el sistema
iniciar_sistema()
```
